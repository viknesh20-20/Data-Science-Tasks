Use the diabetes data set from UCI data set for performing the following:
Apply Bivariate analysis:  Multiple Regression analysis

```python
import pandas as pd
import statsmodels.api as sm
from sklearn.datasets import load_diabetes

# Load the diabetes dataset from scikit-learn
diabetes = load_diabetes()

# Create a DataFrame from the dataset
df_diabetes = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df_diabetes['target'] = diabetes.target

# Bivariate analysis using multiple regression
X = df_diabetes.drop('target', axis=1)  # Independent variables
y = df_diabetes['target']  # Dependent variable

# Add a constant term to the independent variables matrix
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Display the regression summary
print(model.summary())
```

```python
# Create a 5-dimensional array
five_dimensional_array = np.ndarray(shape=(2, 3, 4, 5, 6), dtype=float)  # Adjust the shape as needed

# Verify the number of dimensions
number_of_dimensions = five_dimensional_array.ndim

# Display the array and its dimensions
print("5-Dimensional Array:")
print(five_dimensional_array)

print("\nNumber of Dimensions:", number_of_dimensions)
```

```python
# Create a boolean array
boolean_array = np.array([True, False, True, False, True])

# Sort the boolean array
sorted_boolean_array = np.sort(boolean_array)

# Display the original and sorted boolean arrays
print("Original Boolean Array:")
print(boolean_array)

print("\nSorted Boolean Array:")
print(sorted_boolean_array)
```

i. Write a NumPy program to perform the basic arithmetic operations

ii.Write a NumPy program to transpose an array

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])
# Addition
result_addition = array1 + array2
# Subtraction
result_subtraction = array1 - array2
# Multiplication
result_multiplication = array1 * array2
# Division
result_division = array1 / array2
# Display the results
print("Array 1:")
print(array1)
print("\nArray 2:")
print(array2)
print("\nAddition:")
print(result_addition)
print("\nSubtraction:")
print(result_subtraction)
print("\nMultiplication:")
print(result_multiplication)
print("\nDivision:")
print(result_division)
```

```python
# Create an array for demonstration
original_array = np.array([[1, 2, 3], [4, 5, 6]])

# Transpose the array
transposed_array = np.transpose(original_array)

# Display the original and transposed arrays
print("Original Array:")
print(original_array)

print("\nTransposed Array:")
print(transposed_array)
```
